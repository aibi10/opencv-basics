```python
font = cv2.FONT_HERSHEY_PLAIN

rectange_draw = cv2.rectangle(img, (520, 200), (650, 250), (0, 0, 255), 3)
cv2.putText(rectange_draw, "Rectangle", (520 + 5, 200 - 10), font,  1, (0, 0, 255), 2)

cv2.imshow("rectange_draw", rectange_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
